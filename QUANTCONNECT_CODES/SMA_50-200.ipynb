```python
from AlgorithmImports import *

class SimpleMovingAverageAlgorithm(QCAlgorithm):
    def Initialize(self):
        self.SetStartDate(2021, 1, 1)
        self.SetEndDate(2022, 1, 1)
        self.SetCash(100000)       
        self.symbol = self.AddEquity("SPY", Resolution.Daily).Symbol        
        self.short_window = 50
        self.long_window = 200        
        self.short_sma = self.SMA(self.symbol, self.short_window, Resolution.Daily)
        self.long_sma = self.SMA(self.symbol, self.long_window, Resolution.Daily)
        self.SetWarmUp(self.long_window)
        
    def OnData(self, data):
        if self.IsWarmingUp:
            return
        
        if not self.Portfolio.Invested and self.short_sma.Current.Value > self.long_sma.Current.Value:
            self.SetHoldings(self.symbol, 1)  
        elif self.Portfolio.Invested and self.short_sma.Current.Value < self.long_sma.Current.Value:
            self.Liquidate(self.symbol) 
```
