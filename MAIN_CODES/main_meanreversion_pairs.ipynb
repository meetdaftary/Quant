**STRATEGY 1: PAIR TRADING FOR ADANIENT.NS AND ADANIPORTS.NS (HIGH CORRELATION)**

```python
import numpy as np

pairtrade = PairTrading()
portfolio_values = pairtrade.backtest_trade1(S1 = stock_c1['ADANIENT.NS'], S2 = stock_c1['ADANIPORTS.NS'], money = 100000, scale_factor = 0.05, window1=5, window2=20)
print(portfolio_values)

backtest= BacktestMetrics()
portfolio_values = np.array(portfolio_values)
portfolio_returns = np.diff(portfolio_values) / portfolio_values[:-1]

strategy_cagr = backtest.calculate_cagr(start_value = portfolio_values[0], end_value = portfolio_values[-1], years = (stock_c1['ADANIENT.NS'].index[-1] - stock_c1['ADANIENT.NS'].index[0]).days / 365)
strategy_sharpe = backtest.calculate_sharpe_ratio(returns = portfolio_returns)
print("STRATEGY CAGR:", strategy_cagr, "%")
print("STRATEGY SHARPE:", strategy_sharpe)

benchmark_returns = stock_c1['^NSEI'].pct_change().dropna()
benchmark_cagr = backtest.calculate_cagr(stock_c1['^NSEI'][0], stock_c1['^NSEI'][-1], (stock_c1['ADANIENT.NS'].index[-1] - stock_c1['ADANIENT.NS'].index[0]).days / 365)
benchmark_sharpe = backtest.calculate_sharpe_ratio(benchmark_returns)
print("BENCHMARK CAGR:", benchmark_cagr, "%")
print("BENCHMARK SHARPE:", benchmark_sharpe)
```

**STRATEGY 3: ETF arbitrage takes advantage of price discrepancies between an exchange-traded fund (ETF) and its underlying assets. When the market price of an ETF deviates from the value of its net asset value (NAV), we can exploit the difference. If an ETF trades at a premium (higher than NAV), we can sell the ETF and buy the underlying assets, profiut is the spread. Conversely, if the ETF trades at a discount (lower than its NAV), we can buy the ETF and sell the underlying assets. This strategy will help us to pocket the difference due to mispricings and will always be profitable in any market conditions.**
