**STRATEGY 1: PAIR TRADING FOR BRITANNIA.NS AND ONGC.NS (LOW CORRELATION)**

```python
momentum_trade = Momentum()
portfolio_values = momentum_trade.backtest_trade(stock = stock_c1, x= 'BRITANNIA.NS', y = 'ONGC.NS', z = 30)
print(portfolio_values)

backtest= BacktestMetrics()
portfolio_values = np.array(portfolio_values)
portfolio_returns = np.diff(portfolio_values) / portfolio_values[:-1]

strategy_cagr = backtest.calculate_cagr(start_value = portfolio_values[0], end_value = portfolio_values[-1], years = (stock_c['ADANIENT.NS'].index[-1] - stock_c['ADANIENT.NS'].index[0]).days / 365)
strategy_sharpe = backtest.calculate_sharpe_ratio(returns = portfolio_returns)
print("STRATEGY CAGR:", strategy_cagr, "%")
print("STRATEGY SHARPE:", strategy_sharpe)

benchmark_returns = stock_c['^NSEI'].pct_change().dropna()
benchmark_cagr = backtest.calculate_cagr(stock_c['^NSEI'][0], stock_c['^NSEI'][-1], (stock_c['ADANIENT.NS'].index[-1] - stock_c['ADANIENT.NS'].index[0]).days / 365)
benchmark_sharpe = backtest.calculate_sharpe_ratio(benchmark_returns)
print("BENCHMARK CAGR:", benchmark_cagr, "%")
print("BENCHMARK SHARPE:", benchmark_sharpe)
```
